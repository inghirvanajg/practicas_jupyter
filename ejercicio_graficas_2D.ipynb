## Ejercicio 2

Graficar la funci√≥n x^2 desde x=-5 hasta x=5

```python
x = np.linspace(-5,5,100) 
y = x ** 2
plt.plot(x,y, color = 'pink')
plt.xlabel('x')
plt.ylabel('y = $x^2$')
plt.title('Parabola')
plt.show()
```

```python
x = np.linspace(-5,5,100) 
y = x ** 2
plt.plot(x,y, 'm.')
plt.xlabel('x')
plt.ylabel('y = $x^2$')
plt.title('Parabola')
plt.show()
```

```python
pi = np.pi
x = np.arange(-2*pi,2*pi,0.01)
y = np.sin(x)/x
y1 = np.sin(2*x)/x
y2 = np.sin(3*x)/x
plt.plot(x,y, 'k--', linewidth=3)
plt.plot(x,y1, 'r-.')
plt.plot(x,y2)
plt.show()
```

```python
pi = np.pi
x = np.arange(-2*pi,2*pi,0.01)
y = np.sin(x)/x
y1 = np.sin(2*x)/x
y2 = np.sin(3*x)/x
plt.plot(x,y, 'k--', linewidth=3)
plt.plot(x,y1, 'r-.')
plt.plot(x,y2)
plt.xlabel('x')
plt.ylabel('funciones')
plt.legend(["y","y1","y2"])
plt.xticks([-2*pi, -pi, 0, pi, 2*pi],['r$-2\pi$','r$\pi$','r$0$','r$\pi$','r$2\pi$'])
plt.show()
```
